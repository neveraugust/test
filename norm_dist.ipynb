```python
import math
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
u = 0   # 均值μ
sig = math.sqrt(1)  # 标准差δ
x = np.linspace(u - 4*sig, u + 4*sig, 100000)   # 定义域
y = np.exp(-(x - u) ** 2 / (2 * sig ** 2)) / (math.sqrt(2*math.pi)*sig) # 定义曲线函数
plt.plot(x, y, "black", linewidth=1.5)    # 加载曲线
plt.grid(True)  # 网格线
plt.fill_between(x,0,y,where = abs(x)>=2,facecolor="grey",alpha = 0.7)
#plt.axis('off')
plt.show() 
```
